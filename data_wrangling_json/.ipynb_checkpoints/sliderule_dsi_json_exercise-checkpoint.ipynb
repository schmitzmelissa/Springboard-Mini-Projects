# 10 Countries with Most Projects
Since the countyname column is not nested, we can perform a value count directly on the column.

```python
# Count total count for each country
projects_per_country = df.countryname.value_counts()


# Pull only the top 10 values
top_10_project_counts = pd.DataFrame(projects_per_country.head(10))
top_10_project_counts
```
